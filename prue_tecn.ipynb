```python
# d. ¿Qué palabras son más frecuentes?

from collections import Counter

lista_tweets = tweet_text.text.to_list()
new_lista_tweets = []

for tweet in lista_tweets:
    palabras = tweet.split()
    new_lista_tweets += palabras

contador = Counter(new_lista_tweets)
top_mas_repetidas = contador.most_common(15)

print(top_mas_repetidas[3], top_mas_repetidas[7], top_mas_repetidas[11])

# (Ver NLTK)
```
